```python
import pandas as pd
```

```python
df = pd.read_csv('../data/clubs_with_interest_areas.csv')
```

```python
df.interest_areas.value_counts()
```

```python
interest_map = {"Sports & Fitness" : 1,
"Culture & Languages" : 2,
"Science & Research" : 3,
"Debate & Public Speaking" : 4,
"Business & Finance" : 5,
"Technology & Computing" : 6,
"Music & Performance" : 7,
"Arts & Creativity" : 8,
"Social Impact & Volunteering" : 9,
"Gaming & Esports" : 10,
"Dance" : 11,
"Film & Media" : 12 }
```

```python
def map_interests(text):
    items = text.split('; ')
    return [interest_map[item] for item in items if item in interest_map]

df['interest_areas_ids'] = df['interest_areas'].apply(map_interests)

print(df[['interest_areas', 'interest_areas_ids']].head())
```
